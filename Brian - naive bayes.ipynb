# Adding imports

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics 
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

# Adding adverage goals scoredand conceded for home and away team

```python
allPremDataDF = pd.read_csv("csv/FullPremierLeagueDetails.csv")
allPremDataDF[['homeGoalsScored', 'homeGolsConceded']] = allPremDataDF.homeGoals.str.split(":", expand = True,)
allPremDataDF[['awayGoalsScored', 'awayGolsConceded']] = allPremDataDF.awayGoals.str.split(":", expand = True,)
# allPremDataDF
cleanedallPremDataDF = allPremDataDF.drop(['homeGoals'], axis=1).drop(['awayGoals'], axis=1).drop(['homeGoalDif'], axis=1).drop(['awayGoalDif'], axis=1)
cleanedallPremDataDF
```

# Setting the average goals scored and conceded for home and away team

```python
cleanedallPremDataDF['homeGoalsScored'] = (pd.to_numeric(cleanedallPremDataDF['homeGoalsScored']) / pd.to_numeric(cleanedallPremDataDF['homePlayed'])).round(2)
cleanedallPremDataDF['homeGolsConceded'] = (pd.to_numeric(cleanedallPremDataDF['homeGolsConceded']) / pd.to_numeric(cleanedallPremDataDF['homePlayed'])).round(2)
cleanedallPremDataDF['awayGoalsScored'] = (pd.to_numeric(cleanedallPremDataDF['awayGoalsScored']) / pd.to_numeric(cleanedallPremDataDF['awayPlayed'])).round(2)
cleanedallPremDataDF['awayGolsConceded'] = (pd.to_numeric(cleanedallPremDataDF['awayGolsConceded']) / pd.to_numeric(cleanedallPremDataDF['awayPlayed'])).round(2)

# averageGoalsAtHome = cleanedallPremDataDF['homeGoalsScored'].mean()
# averageGoalsAway = cleanedallPremDataDF['awayGoalsScored'].mean()
# averageGoalsConcededAtHome = cleanedallPremDataDF['homeGolsConceded'].mean()
# averageGoalsConcededAway = cleanedallPremDataDF['awayGolsConceded'].mean()

# print(averageGoalsAtHome)
# print(averageGoalsAway)
# print(averageGoalsConcededAtHome)
# print(averageGoalsConcededAway)
cleanedallPremDataDF
```

```python
# cleanedallPremDataDF['HomeAttackRate'] = 
#             (cleanedallPremDataDF['homeGoalsScored'] / 
#              (cleanedallPremDataDF['homePlayed'] / 2)) 
#                 / averageGoalsAtHome
# cleanedallPremDataDF['HomeDefenseRate'] = (cleanedallPremDataDF['homeGolsConceded'] / (cleanedallPremDataDF['homePlayed'] / 2)) / averageGoalsConcededAtHome
# cleanedallPremDataDF['AwayAttackRate'] = (cleanedallPremDataDF['awayGoalsScored'] / (cleanedallPremDataDF['awayPlayed'] / 2)) / averageGoalsAway
# cleanedallPremDataDF['AwayDefenseRate'] = (cleanedallPremDataDF['awayGolsConceded'] / (cleanedallPremDataDF['awayPlayed'] / 2)) / averageGoalsConcededAway
cleanedallPremDataDF['HomeLiklyGoals'] = cleanedallPremDataDF['homeGoalsScored'] * cleanedallPremDataDF['awayGolsConceded']
cleanedallPremDataDF['AwayLiklyGoals'] = cleanedallPremDataDF['awayGoalsScored'] * cleanedallPremDataDF['homeGolsConceded']
cleanedallPremDataDF
```

# Getting averages per game

```python
cleanedallPremDataDF['homeWon'] = pd.to_numeric(cleanedallPremDataDF['homeWon']) / pd.to_numeric(cleanedallPremDataDF['homePlayed'])
cleanedallPremDataDF['homeDraw'] = pd.to_numeric(cleanedallPremDataDF['homeDraw']) / pd.to_numeric(cleanedallPremDataDF['homePlayed'])
cleanedallPremDataDF['homeLost'] = pd.to_numeric(cleanedallPremDataDF['homeLost']) / pd.to_numeric(cleanedallPremDataDF['homePlayed'])
cleanedallPremDataDF['homePoints'] = pd.to_numeric(cleanedallPremDataDF['homePoints']) / pd.to_numeric(cleanedallPremDataDF['homePlayed'])

cleanedallPremDataDF['awayWon'] = pd.to_numeric(cleanedallPremDataDF['awayWon']) / pd.to_numeric(cleanedallPremDataDF['awayPlayed'])
cleanedallPremDataDF['awayDraw'] = pd.to_numeric(cleanedallPremDataDF['awayDraw']) / pd.to_numeric(cleanedallPremDataDF['awayPlayed'])
cleanedallPremDataDF['awayLost'] = pd.to_numeric(cleanedallPremDataDF['awayLost']) / pd.to_numeric(cleanedallPremDataDF['awayPlayed'])
cleanedallPremDataDF['awayPoints'] = pd.to_numeric(cleanedallPremDataDF['awayPoints']) / pd.to_numeric(cleanedallPremDataDF['awayPlayed'])
cleanedallPremDataDF = cleanedallPremDataDF.drop(['homePlayed'], axis=1).drop(['awayPlayed'], axis=1)
cleanedallPremDataDF
```

# Getting only numeric values from the dataframe

```python
cleanedallPremDataDF = cleanedallPremDataDF.apply (pd.to_numeric, errors='coerce')
cleanedallPremDataDF = cleanedallPremDataDF.dropna()
```

# Setting up score to be the result field and using 25% of the data for testing

```python
#x = cleanedallPremDataDF.drop('score', axis=1)
#x = cleanedallPremDataDF[['homePosition','awayPosition']]
#x = cleanedallPremDataDF[['homeGoalsScored','homeGolsConceded', 'awayGoalsScored', 'awayGolsConceded', 'homePosition', 'awayPosition']]
x = cleanedallPremDataDF[['HomeLiklyGoals', 'AwayLiklyGoals']]
y = cleanedallPremDataDF['score']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 42)
print("{0:0.2f}% in training set".format((len(x_train)/len(cleanedallPremDataDF.index)) * 100))
print("{0:0.2f}% in test set".format((len(x_test)/len(cleanedallPremDataDF.index)) * 100))
```

```python
print("Original Over : {0} ({1:0.2f}%)".format(len(cleanedallPremDataDF.loc[cleanedallPremDataDF['score'] == 1]), (len(cleanedallPremDataDF.loc[cleanedallPremDataDF['score'] == 1])/len(cleanedallPremDataDF.index)) * 100))
print("Original Under : {0} ({1:0.2f}%)".format(len(cleanedallPremDataDF.loc[cleanedallPremDataDF['score'] == 0]), (len(cleanedallPremDataDF.loc[cleanedallPremDataDF['score'] == 0])/len(cleanedallPremDataDF.index)) * 100))
print("")
print("Training Over : {0} ({1:0.2f}%)".format(len(y_train[y_train[:] == 1]), (len(y_train[y_train[:] == 1])/len(y_train)) * 100))
print("Trianing Under : {0} ({1:0.2f}%)".format(len(y_train[y_train[:] == 0]), (len(y_train[y_train[:] == 0])/len(y_train)) * 100))
print("")
print("Test Over : {0} ({1:0.2f}%)".format(len(y_test[y_test[:] == 1]), (len(y_test[y_test[:] == 1])/len(y_test)) * 100))
print("Test Under : {0} ({1:0.2f}%)".format(len(y_test[y_test[:] == 0]), (len(y_test[y_test[:] == 0])/len(y_test)) * 100))
print("")
```

# Testing Accuracy

```python
model = GaussianNB()
model.fit(x_train, y_train)
```

```python
y_pred_train = model.predict(x_train)
y_pred_test = model.predict(x_test)
print("Train Data Accuracy: {}%".format(round(np.mean(y_pred_train == y_train)*100)))
print("Test Data Accuracy: {}%".format(round(np.mean(y_pred_test == y_test)*100)))
```
