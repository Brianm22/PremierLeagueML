```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
from sklearn import metrics 
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn import linear_model
from sklearn import svm
```

```python

allPremDataDF = pd.read_csv("csv/FullPremierLeagueDetails.csv")
allPremDataDF[['homeGoalsScored', 'homeGolsConceded']] = allPremDataDF.homeGoals.str.split(":", expand = True,)
allPremDataDF[['awayGoalsScored', 'awayGolsConceded']] = allPremDataDF.awayGoals.str.split(":", expand = True,)
cleanedallPremDataDF = allPremDataDF.drop(['homeGoals'], axis=1).drop(['awayGoals'], axis=1).drop(['homeGoalDif'], axis=1).drop(['awayGoalDif'], axis=1)

cleanedallPremDataDF['homeGoalsScored'] = pd.to_numeric(cleanedallPremDataDF['homeGoalsScored']) / pd.to_numeric(cleanedallPremDataDF['homePlayed'])
cleanedallPremDataDF['homeGolsConceded'] = pd.to_numeric(cleanedallPremDataDF['homeGolsConceded']) / pd.to_numeric(cleanedallPremDataDF['homePlayed'])
cleanedallPremDataDF['awayGoalsScored'] = pd.to_numeric(cleanedallPremDataDF['awayGoalsScored']) / pd.to_numeric(cleanedallPremDataDF['awayPlayed'])
cleanedallPremDataDF['awayGolsConceded'] = pd.to_numeric(cleanedallPremDataDF['awayGolsConceded']) / pd.to_numeric(cleanedallPremDataDF['awayPlayed'])

cleanedallPremDataDF['homeWon'] = pd.to_numeric(cleanedallPremDataDF['homeWon']) / pd.to_numeric(cleanedallPremDataDF['homePlayed'])
cleanedallPremDataDF['homeDraw'] = pd.to_numeric(cleanedallPremDataDF['homeDraw']) / pd.to_numeric(cleanedallPremDataDF['homePlayed'])
cleanedallPremDataDF['homeLost'] = pd.to_numeric(cleanedallPremDataDF['homeLost']) / pd.to_numeric(cleanedallPremDataDF['homePlayed'])
cleanedallPremDataDF['homePoints'] = pd.to_numeric(cleanedallPremDataDF['homePoints']) / pd.to_numeric(cleanedallPremDataDF['homePlayed'])

cleanedallPremDataDF['awayWon'] = pd.to_numeric(cleanedallPremDataDF['awayWon']) / pd.to_numeric(cleanedallPremDataDF['awayPlayed'])
cleanedallPremDataDF['awayDraw'] = pd.to_numeric(cleanedallPremDataDF['awayDraw']) / pd.to_numeric(cleanedallPremDataDF['awayPlayed'])
cleanedallPremDataDF['awayLost'] = pd.to_numeric(cleanedallPremDataDF['awayLost']) / pd.to_numeric(cleanedallPremDataDF['awayPlayed'])
cleanedallPremDataDF['awayPoints'] = pd.to_numeric(cleanedallPremDataDF['awayPoints']) / pd.to_numeric(cleanedallPremDataDF['awayPlayed'])
cleanedallPremDataDF = cleanedallPremDataDF.drop(['homePlayed'], axis=1).drop(['awayPlayed'], axis=1)
cleanedallPremDataDF.head()
```

```python
cleanedallPremDataDF = cleanedallPremDataDF.apply (pd.to_numeric, errors='coerce')
cleanedallPremDataDF = cleanedallPremDataDF.dropna()
```

```python
x = cleanedallPremDataDF.drop('score', axis=1)
y = cleanedallPremDataDF['score']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
cls = KNeighborsClassifier()
naive_bais = GaussianNB()
lr_model = LogisticRegression(C=0.7, random_state=42, max_iter=1000)
svcl = svm.SVC(kernel='linear')
svcp = svm.SVC(kernel='poly',degree=3)
svcrbf = svm.SVC(kernel='rbf')
```

```python
print('Naive bayes')
naive_bais.fit(x_train, y_train.ravel())
nb_y_pred_train = naive_bais.predict(x_train)
nb_y_pred_test = naive_bais.predict(x_test)
print("Train Data Accuracy: {}%".format(round(np.mean(nb_y_pred_train == y_train)*100)))
print("Test Data Accuracy: {}%".format(round(np.mean(nb_y_pred_test == y_test)*100)))

print('********************************************')
print('Nearest neighbour')
cls.fit(x_train, y_train.ravel())
nn_y_pred_train = cls.predict(x_train)
nn_y_pred_test = cls.predict(x_test)
print("Train Data Accuracy: {}%".format(round(np.mean(nn_y_pred_train == y_train)*100)))
print("Test Data Accuracy: {}%".format(round(np.mean(nn_y_pred_test == y_test)*100)))

print('********************************************')
print('Logistic Regression')
lr_model.fit(x_train, y_train.ravel())
lr_y_pred_train = lr_model.predict(x_train)
lr_y_pred_test = lr_model.predict(x_test)
print("Train Data Accuracy: {}%".format(round(np.mean(lr_y_pred_train == y_train)*100)))
print("Test Data Accuracy: {}%".format(round(np.mean(lr_y_pred_test == y_test)*100)))

print('********************************************')
print('Support Vector Machine Linear')
svcl.fit(x_train, y_train.ravel())
svcl_y_pred_train = svcl.predict(x_train)
svcl_y_pred_test = svcl.predict(x_test)
print("Train Data Accuracy: {}%".format(round(np.mean(svcl_y_pred_train == y_train)*100)))
print("Test Data Accuracy: {}%".format(round(np.mean(svcl_y_pred_test == y_test)*100)))

print('********************************************')
print('Support Vector Machine Polynomial')
svcp.fit(x_train, y_train.ravel())
svcp_y_pred_train = svcp.predict(x_train)
svcp_y_pred_test = svcp.predict(x_test)
print("Train Data Accuracy: {}%".format(round(np.mean(svcp_y_pred_train == y_train)*100)))
print("Test Data Accuracy: {}%".format(round(np.mean(svcp_y_pred_test == y_test)*100)))

print('********************************************')
print('Support Vector Machine RBF')
svcrbf.fit(x_train, y_train.ravel())
svcrbf_y_pred_train = svcrbf.predict(x_train)
svcrbf_y_pred_test = svcrbf.predict(x_test)
print("Train Data Accuracy: {}%".format(round(np.mean(svcrbf_y_pred_train == y_train)*100)))
print("Test Data Accuracy: {}%".format(round(np.mean(svcrbf_y_pred_test == y_test)*100)))
```
